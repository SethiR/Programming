__Updating data__

```python
people = {
    "first" : ["Sam", "Pete", "Sara"],
    "last" : ["Nelson", "Nelson","Green"],
    "email" : ["sam@gmail.com", "pete@gmail.com","sara@hotmail.com"],
}
dfs = pd.DataFrame(people)
dfs
```

```python
dfs.columns = ['first_name', 'last_name', 'email']
dfs
```

```python
dfs.columns = [x.upper() for x in dfs.columns]
dfs
```

```python
dfs.columns = ['first name', 'last name', 'email']
dfs
```

```python
dfs.columns = dfs.columns.str.replace(' ','_')
dfs
```

```python
dfs.rename(columns= {'first_name' : 'first'}, inplace=True)
dfs
```

```python
dfs.rename(columns= {'last_name' : 'last'}, inplace=True)
dfs
```

```python
# Updating values for all cols of a row.
dfs.loc[2] = ['Sara', 'Green', 'sara@yahoomail.com']
dfs
```

```python
# Updating values for certain cols of a row.
dfs.loc[2, ['first', 'email']] = ['Sarah', 'sarah@yahoomail.com']
dfs
```

```python
dfs.loc[2, 'email'] =  'sarah@live.com'
dfs
```

```python
dfs.at[2, 'email'] =  'sarah@zoho.com'   # .at Access a single value for a row/column label pair.
dfs
```

Updating multiple rows of data.

```python
dfs['email'] = dfs.loc[:,'email'].str.upper()
dfs
```
